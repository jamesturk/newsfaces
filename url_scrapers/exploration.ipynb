```python
# Import necessary libraries
import time
import requests
import urllib
import lxml.html
from bs4 import BeautifulSoup
```

```python
# CNN
resp = requests.get(
    "https://edition.cnn.com/2023/06/20/politics/trump-indictment-cabinet-secretaries-what-matters/index.html"
)

root = lxml.html.fromstring(resp.text)
main_article = root.cssselect("div.article__content-container")
# print(main_article[0].text_content())
paragraphs = root.cssselect("p.paragraph.inline-placeholder")
for p in paragraphs:
    print(p.text_content())
```

```python
# Fox News

resp = requests.get(
    "https://www.foxnews.com/media/anti-woke-movie-guide-site-gives-conservatives-alternative-rotten-tomatoes"
)
root = lxml.html.fromstring(resp.text)
main_article = root.cssselect("div.article-body")
paragraphs = main_article[0].cssselect("p")

for p in paragraphs:
    print(p.text_content())
```

```python
# Washington post

resp = requests.get(
    "https://www.washingtonpost.com/politics/2023/06/21/schiff-censure-house-vote/"
)
root = lxml.html.fromstring(resp.text)
main_article = root.cssselect("div.grid-body")[0].cssselect("p")
paragraphs = root.cssselect("div.article-body.p")
print(main_article[0].text_content())
for p in main_article:
    print(p.text_content())
```

```python
# NPR
resp = requests.get(
    "https://www.npr.org/2023/06/21/1183383477/trump-gop-primary-2024-biden-approval-rating-democrats"
)
root = lxml.html.fromstring(resp.text)
main_article = root.cssselect("div.storytext")[0].cssselect("p")
for p in main_article:
    print(p.text_content(), "\n")
```
