```python
def obtain_page_urls(section="senate",page="1"):
    """
    Obtain the urls for a page of results from the politics section of Breitbart
    Inputs:
    -page(str): page of results to search in results
    Return:
    links_list(list): List of urls 
    """
    url = "https://thehill.com/homenews/{}/?page_num={}".format(section,page)
    resp = requests.get(url)
    root = lxml.html.fromstring(resp.text)
    articles_elements = root.cssselect("div.archive__item__content")
    links_list = []
    for article in articles_elements:
        url = article.cssselect("a")[0].get("href")
        #full_link = "breitbart.com" + str(url)
        links_list.append(url)
    
    return links_list

obtain_page_urls()
```
