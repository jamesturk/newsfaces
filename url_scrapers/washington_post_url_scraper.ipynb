```python
#Import necessary libraries
import time
import requests
import urllib
import lxml.html
from bs4 import BeautifulSoup
import newspaper
import trafilatura
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
def find_link_elements(driver):
    url_set = set()
    articles = driver.find_elements(By.CSS_SELECTOR,'div.story-headline')
    for article in articles:
        url_set.add(article.find_elements(By.CSS_SELECTOR,'a.hover-blue')[0].get_attribute("href"))

    return url_set
```

```python
def test_scrapers_wp(max_pages_to_crawl=10):
    links_list = set()
    driver = webdriver.Chrome()
    driver.get("https://www.washingtonpost.com/politics/")
    for i in range(max_pages_to_crawl):
        print("obtaining links from page:", i)
        time.sleep(2)
        links_list.update(find_link_elements(driver))
        #driver.find_elements(By.CSS_SELECTOR,"div.pagination-arrow-right")
        driver.find_element(By.CSS_SELECTOR,"button.inline-flex").click()
        time.sleep(2)
        #next_button[0].click()

    return links_list

```

```python
def test_scrapers_wp_2(max_pages_to_crawl=10):
    links_list = set()
    driver = webdriver.Chrome()
    driver.get("https://www.washingtonpost.com/politics/")
    for i in range(max_pages_to_crawl):
        print("obtaining links from page:", i)
        time.sleep(2)
        driver.find_element(By.CSS_SELECTOR,"button.inline-flex").click()
        time.sleep(2)
    links_list.update(find_link_elements(driver))

    return links_list
```
