```python
def get_urls(url, session=None):
    """
    This function takes a URLs and returns lists of URLs
    for containing each article on that page.

    Parameters:
        * url:  a URL to a page of articles
        * session: optional session object parameter
        * selectors: a list of css selectors

    Returns:
        A list of article URLs on that page.
    """
    response = page_grab(url, session)
    urls = []
    selectors=['div.FourColumnContainer-column','div.TwoColumnContainer7030','div.PageList-items']
    for a in selectors:
        container=response.cssselect(a)
        for j in container[0]:
            atr = j.cssselect("a")
            for a in atr:
                print(a)
                href = a.get("href")
                print(href)
                if href is not None:
                    urls.append(href)
    return urls

get_urls('https://apnews.com/politics', session=None)
```
